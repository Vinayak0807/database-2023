1. **<u>Retrieve the total number of games played by each player.</u>**

```sql
SELECT Name, SUM(Games) AS TotalGames
FROM baseball
GROUP BY Name;
```

1. **<u>Display the player with the highest number of home runs.</u>**

```sql
SELECT Name, Homers
FROM baseball
WHERE Homers = (SELECT MAX(Homers) FROM baseball);
```

1. **<u>Find the player with the maximum at-bats.</u>**

```sql
SELECT Name, AVG(AtBats)
from baseball
group by Name
order by AVG(AtBats) desc
LIMIT 5;
```

1. **<u>Show the player with the most triples.</u>**

1. **<u>Display the names of players who have hit at least 20 home runs and have more than 80 RBIs.</u>**

1. **<u>Show the players with more than 10 doubles and a batting average greater than 0.300.......?</u>**

```sql
SELECT Name , Doubles ,AVG(AtBats) as avg_run
from baseball
GROUP by Name ,GROUP BY Doubles
having Doubles >=10;
```

1. **<u>List the players with more than 25 doubles and an average of at least 3 RBIs per game.</u>**

```sql
SELECT Name, SUM(Doubles) AS Doubles, AVG(RBIs) AS avgrb
FROM baseball
GROUP BY Name
HAVING SUM(Doubles) > 25 AND AVG(RBIs) >= 3;
```
