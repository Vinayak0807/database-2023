```sql
--- innner join for city and country
SELECT
     city.city AS city,
     country.country AS country
FROM city
INNER JOIN country ON city.country_id = country.country_id
ORDER BY city.city;  -- Specify the column for sorting
    
```

```sql
SELECT
    address.address,
    city.city,
    address.district,
    address.phone
from address inner join city
on address.city_id = city.city_id
order BY
    city.city    
```

```sql
SELECT customer.first_name, 
       address.address_id
FROM customer
INNER JOIN address ON customer.address_id = address.address_id;

```
